## Representação e visualização de imagem

Uma imagem digital nada mais é que uma uma matriz de linhas e colunas, onde cada posição desta matriz contem o valor de um *pixel*.

O valor de cada pixel representa a intensidade de cor naquele ponto especifico. 


````
# Importando a biblioteca OpenCV
import cv2 

#linha magica para imprimir graficos no notebook
%matplotlib inline

# Importe da biblioteca para exibir gráficos 
from matplotlib import pyplot as plt


print ("OpenCV Versão : %s " % cv2.__version__)
````

````
# carregar uma imagem
image = cv2.imread("NATUREZA_1.jpg")

# para exibir a imagem
plt.imshow(image)
plt.axis('off') 
plt.show()
````

Mas porque a imagem é mostrada de modo estranho pelo pacote matplotlib? 

R: Porque a OpenCV representa os canais da imagem na ordem B - G - R, e não R - G - B como é esperado pela maior parte das bibliotecas.

Assim, para podermos visualizar corretamente uma imagem do OpenCV com matplotlib, precisamos inverter os canais, como no código abaixo:


````
# converter a imagem de BGR para RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# exibir a imagem
plt.imshow(image_rgb)
plt.axis('off') 
plt.show()
````

````
# Outro método de exibir imagem usando o método imshow da cv2
# Use o ESC para fechara tela

cv2.imshow("img", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

## repare que nesse caso a imagem é exibida em BGR
````

````
# Mostrando a representação interna da imagem
print("Dimensões da imagem: ", image_rgb.shape)
print("Quantidade de linhas: ", image_rgb.shape[0])
print("Quantidade de colunas: ", image_rgb.shape[1])
print("Camadas de cores: ", image_rgb.shape[2])
````

````
# Mostrando a representação interna da imagem
print("Dimensões da imagem: \n", image_rgb)
````

## Amostragem da imagem

As imagens capturadas por dispositivos digitais possuem características como resolução espacial e profundidade de cor.

A resolução espacial define a quantidade de pixels em largura e altura, influenciando a nitidez da imagem. A profundidade de cor determina quantas cores podem ser representadas, dependendo da quantidade de bits alocados para cada canal de cor.

Embora imagens de diferentes tamanhos possam ser comparadas em termos de resolução, é importante considerar a densidade de pixels (PPI/DPI) para uma análise mais precisa.

### Alterando o tamanho de uma imagem

O OpenCV permite redimensionar imagens com a função:

```python
cv2.resize(imagem, (W, H), interpolation=cv2.INTER_LINEAR)
```

Onde (W, H) representa a nova largura e altura, e interpolation define o método de interpolação usado para suavizar ou detalhar a imagem. 

Métodos comuns incluem cv2.INTER_LINEAR, cv2.INTER_CUBIC e cv2.INTER_NEAREST.

Comparar os diferentes métodos de interpolação (vizinho mais próximo, bilinear e bicúbica) ao ampliarmos uma imagem em 10 vezes seu tamanho. Escolha uma imagem pequena.  

Dica de onde encontrar na documentação as flags de interpolação: 
https://docs.opencv.org/3.4/da/d54/group__imgproc__transform.html#ga5bb5a1fea74ea38e1a5445ca803ff121

Para saber mais sobre interpolação, sugiro assistir ao video: https://www.youtube.com/watch?v=8bTDssnJyZc&ab_channel=S.M.RiazulIslam

````
# Carregando a imagem na versão colorida de um arquivo
import cv2
import matplotlib.pyplot as plt
imagem = cv2.imread("img3x3.png")
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

print("Dimensões da imagem original: ", image.shape)

# para testar diferentes métodos de interpolação, vamos redimensionar a imagem para um tamanho maior
# e vamos usar diferentes métodos de interpolação para redimensionar a imagem 
# vamos usar os métodos cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_AREA, cv2.INTER_CUBIC e cv2.INTER_LANCZOS4 
# que são os métodos de interpolação disponíveis no OpenCV
interpolacao = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_AREA, cv2.INTER_CUBIC, cv2.INTER_LANCZOS4]

# essa função é um list comprehension do python que cria uma lista com as imagens redimensionadas com os diferentes métodos de interpolação 
resized_images = [cv2.resize(image, (image.shape[1] * 2, image.shape[0] * 5), interpolation=nome) for nome in interpolacao] 

# Vamos exibir as imagens redimensionadas com os diferentes métodos de interpolação usando o matplotlib com o loop for.
interpolation_names = ['INTER_NEAREST', 'INTER_LINEAR', 'INTER_AREA', 'INTER_CUBIC', 'INTER_LANCZOS4']
plt.figure(figsize=(15, 10))
for i, resized_image in enumerate(resized_images):
    plt.subplot(2, 3, i + 1)
    plt.imshow(resized_image)
    plt.title(interpolation_names[i])
    plt.axis('off')

plt.show()
````

## Alterando os valores dos pixels de uma imagem

### Range de valores

Antes de alterar os valores dos pixels temos que entender que a OpenCV trabalha com valores de **8 bits** para cada componente de cor ou escala de cinza, quer dizer que os valores possiveis estão no range entre **0** e **2⁸-1**, que é a mesma que dizer entre **0 e 255**. 


````
import cv2
import numpy as np


# Carregando a imagem na versão tons de cinza (grayscale) de um arquivo
imagem = cv2.imread("img3x3.png", cv2.IMREAD_COLOR)

# Escolhe um pixel para manipulação.
pixel_original = imagem[1, 2]
print("Valor original do pixel:", pixel_original)

# Alterando o pixel
imagem[1,2] = [255, 120, 50]
print("Novo valor do pixel em (50, 100):", imagem[1, 2])


# Exibindo a imagem
plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
plt.axis('off') 
plt.show()
````

## Varredura de uma imagem 

Implementar uma rotina que percorra a imagem pixel a pixel é uma abordagem didática que facilita o entendimento do processamento de imagens. 

Embora em Python existam métodos mais otimizados para essa tarefa, o uso de dois laços "for" — um para as linhas e outro para as colunas — é uma forma clara e intuitiva de iterar sobre cada elemento da matriz que representa a imagem.

````
import cv2
import matplotlib.pyplot as plt

imagem = cv2.imread("NATUREZA_1.jpg")
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Obtém as dimensões da imagem
altura, largura, _ = imagem_rgb.shape

# Define a linha central e a espessura da faixa a ser alterada
linha_central = altura // 2 
espessura = 10

for y in range(altura):
    for x in range(largura):
      # cria uma condição para verificação
        if y == linha_central:
          imagem_rgb[y-espessura:y+espessura, :] = (255, 0, 0)


plt.imshow(imagem_rgb)
plt.axis("off")
plt.show()
````

# Filtro de imagem

É possivel aplicar diversos tipos de filtros. A filtro abaixo é um filtro de limiarização.

````
import cv2
imagem = cv2.imread("NATUREZA_1.jpg")
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Obtém as dimensões da imagem
altura, largura = image.shape

limiar = 80

for y in range(altura):
  for x in range(largura):
    
    if image[y, x] <= limiar:
      image[y,x] = 0
    else:
      image[y,x] = 255


plt.figure(figsize = (20,20))
plt.subplot(131); plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB));
plt.subplot(132); plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY), cmap="gray");
plt.subplot(133); plt.imshow(image, cmap="gray") 
plt.show()
````
