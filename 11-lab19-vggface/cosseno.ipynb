```python
import numpy as np
import matplotlib.pyplot as plt

def plot_cosine_similarity(vec1, vec2):
    origin = [0, 0]
    
    fig, ax = plt.subplots()
    ax.quiver(*origin, *vec1, angles='xy', scale_units='xy', scale=1, color='r', label='Vetor A')
    ax.quiver(*origin, *vec2, angles='xy', scale_units='xy', scale=1, color='b', label='Vetor B')
    
    # calcular ângulo
    cos_sim = np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
    angle_rad = np.arccos(np.clip(cos_sim, -1.0, 1.0))
    angle_deg = np.degrees(angle_rad)

    ax.set_xlim(-1.5, 2)
    ax.set_ylim(-1.5, 2)
    ax.set_aspect('equal')
    ax.set_title(f"Ângulo: {angle_deg:.2f}° | Similaridade Cosseno: {cos_sim:.2f}")
    ax.grid()
    ax.legend()
    plt.show()

def find_cosine_distance(vec1, vec2):
    a = np.dot(vec1, vec2)
    b = np.linalg.norm(vec1) * np.linalg.norm(vec2)
    return 1 - (a / b)

```

```python
# Exemplos de vetores 2D
v1 = np.array([1, 1])
v2 = np.array([1, 0.5])  # quase mesma direção
plot_cosine_similarity(v1, v2)
find_cosine_distance(v1, v2)
```

```python
# idênticos
v1 = np.array([1, 0])
v2 = np.array([2, 0])
plot_cosine_similarity(v1, v2)


```

```python
# ortogonais
v1 = np.array([1, 0])
v2 = np.array([0, 1])
plot_cosine_similarity(v1, v2)
```

```python
# opostos
v1 = np.array([1, 0])
v2 = np.array([-1, 0])
plot_cosine_similarity(v1, v2)
```
